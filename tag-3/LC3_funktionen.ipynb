Angenommen, wir m√∂chten das noch etwas flexibler halten und vielleicht auch mal die W√∂rter z√§hlen, die 6 Zeichen lang sind ...

```python
# Szenario 1: Mit Defaultwerten 

def count_word_length(text, nr_of_chars=5):

    word_list = []

    text = text.split()
    for word in text:
        if len(word) == nr_of_chars:
            word_list.append(word)

    return len(word_list)
```
